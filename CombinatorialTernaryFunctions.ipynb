```python
#Removes the transformation if the negative of it is also in the set of transformations.
def TrRemover(TRList,p=2):
    #print(len(TRList))
    if p==2:
        remove_list=[]
        for trlist_i1 in range(len(TRList)):
            for trlist_i2 in range(trlist_i1+1,len(TRList)):
                tr1=TRList[trlist_i1]
                tr2=TRList[trlist_i2]
                trch=tr1+tr2
                if all(all(trch[i][j]==0 for j in range(0,2)) for i in range(0,2)):
                #print('trch',trch)
                #print(tr2)
                    if tr2 not in remove_list:
                        remove_list.append(tr2)
        for tr in remove_list:
            TRList.remove(tr)
    #removes trs with determinant not a power of p.
    remove_list2=[]    
    for tr in TRList:
        det_tr=abs(tr.determinant())
        det_factors=factor(det_tr)
        #print(det_tr,det_factors[0][0],len(det_factors),det_factors[0][0])
        if len(det_factors)>1:
            remove_list.append(tr)
        elif len(det_factors)==1 and det_factors[0][0]!=p:
            remove_list2.append(tr)
    for tr in remove_list2:
        TRList.remove(tr)
    return TRList            
```

```python
def TRFinder(k,Q1,Q2,p=2):
    out=[]
    g,h,j,dd,ee,ff=Q1[0],Q1[1],Q1[2],Q1[3],Q1[4],Q1[5]
    a,b,c,d,e,f=Q2[0],Q2[1],Q2[2],Q2[3],Q2[4],Q2[5]
    
    B=a*b*c+1/4*(d*e*f - a*d^2 - b*e^2 - c*f^2)
    lm1=floor(sqrt((b*c-1/4*d^2)*g*k^2/B))
    lm2=floor(sqrt((a*c-1/4*e^2)*g*k^2/B))
    lm3=floor(sqrt((a*b-1/4*f^2)*g*k^2/B))

    lm4=floor(sqrt((b*c-1/4*d^2)*h*k^2/B))
    lm5=floor(sqrt((a*c-1/4*e^2)*h*k^2/B))
    lm6=floor(sqrt((a*b-1/4*f^2)*h*k^2/B))

    lm7=floor(sqrt((b*c-1/4*d^2)*j*k^2/B))
    lm8=floor(sqrt((a*c-1/4*e^2)*j*k^2/B))
    lm9=floor(sqrt((a*b-1/4*f^2)*j*k^2/B))
    
    for l in range(-lm1,lm1+1):
        for r in range(-lm2,lm2+1):
            for u in range(-lm3,lm3+1):
                if a*l^2+b*r^2+c*u^2+d*r*u+e*u*l+f*l*r==k^2*g:
                    for m in range(-lm4,lm4+1):
                        for s in range(-lm5,lm5+1):
                            for v in range(-lm6,lm6+1):
                                if a*m^2+b*s^2+c*v^2+d*s*v+e*v*m+f*m*s==k^2*h:
                                    for n in range(-lm7,lm7+1):
                                        for t in range(-lm8,lm8+1):
                                            for w in range(-lm9,lm9+1):
                                                if a*n^2+b*t^2+c*w^2+d*t*w+e*w*n+f*n*t==k^2*j and\
                                                2*a*l*m+2*b*r*s+ 2*c*u*v +d*r*v+\
                                                d*s*u+e*u*m+e*l*v+f*l*s+f*m*r==k^2*ff and\
                                                2*a*l*n+2*b*r*t+ 2*c*u*w +d*r*w+\
                                                d*t*u+e*u*n+e*l*w+f*l*t+f*n*r==k^2*ee and\
                                                2*a*m*n+2*b*s*t+ 2*c*v*w +d*s*w+\
                                                d*t*v+e*v*n+e*m*w+f*m*t+f*n*s==k^2*dd:
                                                    out.append(matrix([[l/k,m/k,n/k],\
                                                                       [r/k,s/k,t/k],\
                                                                       [u/k,v/k,w/k]]))
    return out #TrRemover(out,p)
    
```

```python
def SetFinder(M,m,Q,prim=1):
    out=[]
    a,b,c,d,e,f=Q[0],Q[1],Q[2],Q[3],Q[4],Q[5]
    for x in range(M):
        for y in range(M):
            for z in range(M):
                if prim!=0 and gcd(M,gcd(x,gcd(y,z)))>1:
                    next
                elif (a*x^2+b*y^2+c*z^2+d*y*z+e*z*x+f*x*y)%M==m:
                    out.append([[x,M],[y,M],[z,M]])
    return out
```

```python
def TrCalc(T,f,p=2):
    xyz=T*matrix([[f[0][0]],[f[1][0]],[f[2][0]]])
    x1,y1,z1=xyz[0][0],xyz[1][0],xyz[2][0]

    Mx=gcd([f[0][1],f[1][1],f[2][1],T[0][0]*f[0][1],T[0][1]*f[1][1],T[0][2]*f[2][1]])
    
    My=gcd([f[0][1],f[1][1],f[2][1],T[1][0]*f[0][1],T[1][1]*f[1][1],T[1][2]*f[2][1]])
    
    Mz=gcd([f[0][1],f[1][1],f[2][1],T[2][0]*f[0][1],T[2][1]*f[1][1],T[2][2]*f[2][1]])
    
    if (Mx not in ZZ) or (My not in ZZ) or (Mz not in ZZ):
        print('Consider increasing the modulo!') 
        #This is new in Feb2024. Does not seem to cause \
        #an issue as the denoms of trs are smaller than the modulus.
        return False
    elif x1 in ZZ and y1 in ZZ and z1 in ZZ:
        if gcd([x1,y1,z1])%p==0:
            return False
        else:
            return [[x1%(Mx),Mx],\
                    [y1%(My),My],\
                    [z1%(Mz),Mz]]
    else:
        return False
```

```python
def TrCalcM(m,M,Q,T,f,p=2):
    out=[]
    ftr1=TrCalc(T,f,p)
    if ftr1:
        for i1 in range(floor(M/ftr1[0][1])+1): 
            for i2 in range(floor(M/ftr1[1][1])+1):
                for i3 in range(floor(M/ftr1[2][1])+1):
                    ftr2=[[(ftr1[0][0] + i1*ftr1[0][1])%M,M],\
                          [(ftr1[1][0] + i2*ftr1[1][1])%M,M],\
                          [(ftr1[2][0] + i3*ftr1[2][1])%M,M]]
                    if (Q[0]*ftr2[0][0]^2+Q[1]*ftr2[1][0]^2+Q[2]*ftr2[2][0]^2\
                        +Q[3]*ftr2[1][0]*ftr2[2][0]+Q[4]*ftr2[2][0]*ftr2[0][0]\
                        +Q[5]*ftr2[0][0]*ftr2[1][0])%M==m:
                        #This portion makes sure an additional set is not introduced \
                        #when the modulo is being changed.
                        if gcd(M,gcd(ftr2[0][0],gcd(ftr2[1][0],ftr2[2][0])))==1:
                            if ftr2 not in out:
                                out.append(ftr2)
        if len(out)==0:
            return False
        else:
            return out
    else:
        return False
```

```python
def TRS_Find(Q1,m1,M1,Q2,k,t1,F1,F2,parts,p=2):
    m2=m1*t1
    M2=M1*t1

    TRS1a=TRFinder(k, Q1, Q2,p)
    
    TRS1=[]
    for T in TRS1a:
        T=T*sqrt(t1)
        TRS1.append(T)
    
    candids1=len(TRS1)
    print(len(TRS1))

    TRS2=TRFinder(k, Q2, Q2,p)
    print(len(TRS2))
    
    candids2=len(TRS2)
    error_place=0 #???

    for tr_counter in range(0,len(TRS1)):
        flag2=0
        tr11=TRS1[tr_counter]
        print('Trying:',tr_counter)
        count2=0
        while count2<candids2^(parts-1):
            iterator2=[]
            count22=count2
            while count22>0:
                iterator2.insert(0,count22%(candids2))
                count22=floor(count22/(candids2))

            while len(iterator2)<parts-1:
                iterator2.insert(0,0)            
            error_place=0
            TRSC=[]
            TRSC.append(tr11)
            for j2 in iterator2:
                TRSC.append(TRS2[j2])
            trll=TRSC[0]
            for i in range(1,len(TRSC)):
                trll=tr_prod(TRSC[i],trll)

            tr_e=tr_inv(trll)

            TRSC.append(tr_e)
            Check=CycleCheck(Q2,m2,M2,F1,F2,TRSC,False,p)
            if Check[0]:
                return [tr for tr in TRSC]
            elif Check[0]==False:
                error_place=Check[1]
                if error_place>0 and error_place<len(iterator2):
                    for i0_iter in range(error_place,len(iterator2)):
                        iterator2[i0_iter]=0
                    iterator2[error_place-1]+=1
                    count2=sum(iterator2[exp_it]*candids2^(len(iterator2)-exp_it-1) for exp_it in range(len(iterator2)))-1
                
                count2=count2+1
                
                if count2>=candids2^(parts-1):
                    flag2=1
                    break
            if flag2==1:
                break
```

```python
def CycleCheck(Q2,m2,M2,F1,F2,TRSC,cycle_print=False,p=2):

    CT=[[] for i in range(0,len(TRSC)-1)]
    for f1 in F1:
        trds=TrCalcM(m2,M2,Q2,TRSC[0],f1,p)
        #print(f1,trds)
        if trds:
            for trd in trds:
                if trd not in CT[0]:
                    CT[0].append(trd)
        else:
            return [False,0]
    for i1 in range(1,len(TRSC)-1):
        for si1 in CT[i1-1]:
            trds=TrCalcM(m2,M2,Q2,TRSC[i1],si1,p)            
            if trds:
                for trd in trds:
                    if all(trd not in CT[ii1] for ii1 in range(i1)):
                        if trd not in CT[i1]:
                            CT[i1].append(trd)
                    else:
                        return [False,i1]
    
    for sl1 in CT[len(CT)-1]:
        sl1trd=TrCalc(TRSC[len(TRSC)-1],sl1,p)
        if sl1trd==False:
            return [False,0]
                
    if any(len(CTs)==0 for CTs in CT):
        return [False,0]
    #Finally check if all F2 is there in CTs.
    F2rem=[ff for ff in F2]
    for il1 in range(0,len(CT)):
        for ffrem in CT[il1]:
            F2rem.remove(ffrem)
    if len(F2rem)==0:
        if cycle_print==True:
            print(CT)
        return [True,True]
    else:
        return [False,0]

    
```

```python
#Finds the subsets of representations modulo M when x is even
def SetFinder_xe(M,m,Q,prim=1):
    out=[]
    a,b,c,d,e,f=Q[0],Q[1],Q[2],Q[3],Q[4],Q[5]
    for x in range(M):
        for y in range(M):
            for z in range(M):
                if prim!=0 and gcd(M,gcd(x,gcd(y,z)))>1:
                    next
                elif (a*x^2+b*y^2+c*z^2+d*y*z+e*z*x+f*x*y)%M==m and x%2==0:
                    out.append([[x,M],[y,M],[z,M]])
    return out
```
